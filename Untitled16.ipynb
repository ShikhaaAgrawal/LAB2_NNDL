# 1. Implement and Visualize Activation Functions:

▪ Step Function

```python
import numpy as np
import matplotlib.pyplot as plt

# Step function implementation
def step_function(x):
    return np.where(x >= 0, 1, 0)

# Visualization
x = np.linspace(-10, 10, 100)
y = step_function(x)

plt.plot(x, y)
plt.title("Step Function")
plt.xlabel("Input")
plt.ylabel("Output")
plt.grid(True)
plt.show()
```

Step Function: Maps inputs greater than or equal to 0 to 1 and less than 0 to 0.



▪ Sigmoid Function (Binary and Bipolar)


```python
# Sigmoid function (Binary) implementation
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# Visualization
y = sigmoid(x)

plt.plot(x, y)
plt.title("Sigmoid Function (Binary)")
plt.xlabel("Input")
plt.ylabel("Output")
plt.grid(True)
plt.show()
```

Sigmoid (Binary): Smooth curve mapping input to a range between 0 and 1.


```python
# Bipolar Sigmoid function implementation
def bipolar_sigmoid(x):
    return (2 / (1 + np.exp(-x))) - 1

# Visualization
y = bipolar_sigmoid(x)

plt.plot(x, y)
plt.title("Bipolar Sigmoid Function")
plt.xlabel("Input")
plt.ylabel("Output")
plt.grid(True)
plt.show()
```

Sigmoid (Bipolar): Smooth curve mapping input to a range between -1 and 1.


▪ Tanh Function


```python
# Tanh function implementation
def tanh_function(x):
    return np.tanh(x)

# Visualization
y = tanh_function(x)

plt.plot(x, y)
plt.title("Tanh Function")
plt.xlabel("Input")
plt.ylabel("Output")
plt.grid(True)
plt.show()
```

Tanh: Similar to the Bipolar Sigmoid, but with a sharper curve, mapping input between -1 and 1.


▪ ReLU Function

```python
# ReLU function implementation
def relu(x):
    return np.maximum(0, x)

# Visualization
y = relu(x)

plt.plot(x, y)
plt.title("ReLU Function")
plt.xlabel("Input")
plt.ylabel("Output")
plt.grid(True)
plt.show()
```

ReLU: Sets all negative values to 0 and passes positive values as-is.

# 2. Implement a Simple Neural Network:

• Create a simple neural network with one hidden layer using each activation
function (sigmoid, tanh, and ReLU).

• Train the network on a binary classification task (e.g., XOR problem) using a
small dataset.

• Compare the performance of the neural network with different activation
functions.

* Sigmoid activation function

```python
import numpy as np

# Define the sigmoid activation function
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# Derivative of the sigmoid function
def sigmoid_derivative(x):
    return x * (1 - x)

# Define the XOR dataset
X = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
y = np.array([[0], [1], [1], [0]])

# Define the neural network architecture
input_size = 2
hidden_size = 1
output_size = 1

# Initialize weights and biases randomly
weights_input_hidden = np.random.randn(input_size, hidden_size)
bias_hidden = np.zeros((1, hidden_size))
weights_hidden_output = np.random.randn(hidden_size, output_size)
bias_output = np.zeros((1, output_size))

# Training parameters
learning_rate = 0.1
epochs = 10000

# Training loop
for epoch in range(epochs):
    # Forward propagation
    hidden_layer_input = np.dot(X, weights_input_hidden) + bias_hidden
    hidden_layer_output = sigmoid(hidden_layer_input)

    output_layer_input = np.dot(hidden_layer_output, weights_hidden_output) + bias_output
    output_layer_output = sigmoid(output_layer_input)

    # Calculate the error
    error = y - output_layer_output

    # Backpropagation
    d_output = error * sigmoid_derivative(output_layer_output)
    error_hidden_layer = d_output.dot(weights_hidden_output.T)
    d_hidden_layer = error_hidden_layer * sigmoid_derivative(hidden_layer_output)

    # Update weights and biases
    weights_hidden_output += hidden_layer_output.T.dot(d_output) * learning_rate
    bias_output += np.sum(d_output, axis=0, keepdims=True) * learning_rate
    weights_input_hidden += X.T.dot(d_hidden_layer) * learning_rate
    bias_hidden += np.sum(d_hidden_layer, axis=0, keepdims=True) * learning_rate

# Make predictions on the training data
hidden_layer_input = np.dot(X, weights_input_hidden) + bias_hidden
hidden_layer_output = sigmoid(hidden_layer_input)
output_layer_input = np.dot(hidden_layer_output, weights_hidden_output) + bias_output
output_layer_output = sigmoid(output_layer_input)

print("Predictions:")
print(output_layer_output)
```

```python


import numpy as np
# * Tanh activation function

# Define the tanh activation function
def tanh_function(x):
    return np.tanh(x)

# Derivative of the tanh function
def tanh_derivative(x):
    return 1 - np.tanh(x)**2

# Define the XOR dataset
X = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
y = np.array([[0], [1], [1], [0]])

# Define the neural network architecture
input_size = 2
hidden_size = 1
output_size = 1

# Initialize weights and biases randomly
weights_input_hidden = np.random.randn(input_size, hidden_size)
bias_hidden = np.zeros((1, hidden_size))
weights_hidden_output = np.random.randn(hidden_size, output_size)
bias_output = np.zeros((1, output_size))

# Training parameters
learning_rate = 0.1
epochs = 10000

# Training loop
for epoch in range(epochs):
    # Forward propagation
    hidden_layer_input = np.dot(X, weights_input_hidden) + bias_hidden
    hidden_layer_output = tanh_function(hidden_layer_input)

    output_layer_input = np.dot(hidden_layer_output, weights_hidden_output) + bias_output
    output_layer_output = tanh_function(output_layer_input)

    # Calculate the error
    error = y - output_layer_output

    # Backpropagation
    d_output = error * tanh_derivative(output_layer_output)
    error_hidden_layer = d_output.dot(weights_hidden_output.T)
    d_hidden_layer = error_hidden_layer * tanh_derivative(hidden_layer_output)

    # Update weights and biases
    weights_hidden_output += hidden_layer_output.T.dot(d_output) * learning_rate
    bias_output += np.sum(d_output, axis=0, keepdims=True) * learning_rate
    weights_input_hidden += X.T.dot(d_hidden_layer) * learning_rate
    bias_hidden += np.sum(d_hidden_layer, axis=0, keepdims=True) * learning_rate

# Make predictions on the training data
hidden_layer_input = np.dot(X, weights_input_hidden) + bias_hidden
hidden_layer_output = tanh_function(hidden_layer_input)
output_layer_input = np.dot(hidden_layer_output, weights_hidden_output) + bias_output
output_layer_output = tanh_function(output_layer_input)

print("Predictions (Tanh):")
print(output_layer_output)
```

* Relu activation function

```python
import numpy as np
# * ReLU activation function

# Define the ReLU activation function
def relu(x):
    return np.maximum(0, x)

# Derivative of the ReLU function
def relu_derivative(x):
    return np.where(x > 0, 1, 0)

# Define the XOR dataset
X = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
y = np.array([[0], [1], [1], [0]])

# Define the neural network architecture
input_size = 2
hidden_size = 1
output_size = 1

# Initialize weights and biases randomly
weights_input_hidden = np.random.randn(input_size, hidden_size)
bias_hidden = np.zeros((1, hidden_size))
weights_hidden_output = np.random.randn(hidden_size, output_size)
bias_output = np.zeros((1, output_size))

# Training parameters
learning_rate = 0.1
epochs = 10000

# Training loop
for epoch in range(epochs):
    # Forward propagation
    hidden_layer_input = np.dot(X, weights_input_hidden) + bias_hidden
    hidden_layer_output = relu(hidden_layer_input)

    output_layer_input = np.dot(hidden_layer_output, weights_hidden_output) + bias_output
    output_layer_output = relu(output_layer_input)

    # Calculate the error
    error = y - output_layer_output

    # Backpropagation
    d_output = error * relu_derivative(output_layer_output)
    error_hidden_layer = d_output.dot(weights_hidden_output.T)
    d_hidden_layer = error_hidden_layer * relu_derivative(hidden_layer_output)

    # Update weights and biases
    weights_hidden_output += hidden_layer_output.T.dot(d_output) * learning_rate
    bias_output += np.sum(d_output, axis=0, keepdims=True) * learning_rate
    weights_input_hidden += X.T.dot(d_hidden_layer) * learning_rate
    bias_hidden += np.sum(d_hidden_layer, axis=0, keepdims=True) * learning_rate

# Make predictions on the training data
hidden_layer_input = np.dot(X, weights_input_hidden) + bias_hidden
hidden_layer_output = relu(hidden_layer_input)
output_layer_input = np.dot(hidden_layer_output, weights_hidden_output) + bias_output
output_layer_output = relu(output_layer_input)

print("Predictions (ReLU):")
print(output_layer_output)
```
